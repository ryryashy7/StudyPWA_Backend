# 1. What is Backend Web Development?

A **backend** is the part of a web application that runs on the server (not in the browser).

The backend is responsible for:
- storing and retrieving data
- sending web pages or data to the browser
- managing user accounts and passwords
- connecting the front-end to a database

If the front-end is *what the user sees*,  
the backend is *the behind-the-scenes engine*.

Backend code does **not** run on the userâ€™s computer.  
It runs on a server and sends responses to the user's browser.

# 3. The Clientâ€“Server Model

When you open a website, your browser is called the **client**.

The server is a computer running backend software that:
1. receives the browserâ€™s request

## Simple diagram

# 4. What is a Web Server? What is Flask?

## Web server software

Software that accepts HTTP requests and sends back responses.

Examples:
- Apache
- Nginx
- Node.js/Express
- **Flask development server** â† we will use this

## Flask

Flask is a **Python web framework**.  
A web framework gives us tools to:
- create routes (pages and API endpoints)
- return HTML or JSON
- connect to databases
- manage requests and responses

Flask lets us write code like:

```python
@app.route('/')
def home():
    return "Hello"
```

## Libraries and Objects

- A **library** is pre-written code that we import and use.
- An **object** holds data and functions together (e.g., Flask `request` object).


# Understanding Flask Routes (`@app.route()`)

When we write:

```python
@app.route('/')
def home():
    return "Hello"
```

We are doing three things:

---

## 1. `@app.route('/')` â€“ This defines a *route*
A **route** is a URL on your website.

- `'/'` means the **home page**
- If the browser visits `http://127.0.0.1:5000/`, this route is triggered

You can think of it as:

> â€œWhen someone visits this URL, run the function below.â€

---

## 2. `def home():` â€“ This is just a normal Python function
When this function runs, Flask will use whatever it **returns** as the webpage response.

---

## 3. `return "Hello"` â€“ Flask sends this text to the browser
This means the browser will literally show:

```
Hello
```

---

## Full Request/Response Flow

```text
Browser â†’ GET request for "/"
Flask â†’ runs home()
home() â†’ returns "Hello"
Flask â†’ sends "Hello" back to browser
Browser â†’ displays Hello
```

This is the foundation of backend web programming.

We will replace `"Hello"` with real pages, and later with JSON from a database.

### âœ”ï¸ 2. What does `__name__` mean?

`__name__` is a special Python variable that stores the **name of the current file**.

When you run a file directly (like `python app.py`), Python sets:

```
__name__ = "__main__"
```

Flask uses this to understand where your program lives.

### âœ”ï¸ 3. Why does Flask need this?

- load static files correctly  
- provide helpful error messages  

### ðŸ‘ Summary


# 7. What is an API?

API = **A**pplication **P**rogramming **I**nterface.
It is a way for the front-end to request **data** from the back-end.

APIs normally return **JSON**, not HTML.

## JSON example

If the browser calls:

```
GET /api/subjects
```

The server might return:

```json
[
  { "id": "subj1", "name": "Maths" },
  { "id": "subj2", "name": "English" }
]
```

## How JavaScript uses JSON

```javascript
fetch('/api/subjects')
  .then(r => r.json())
  .then(data => console.log(data));
```

This is how the browser and backend communicate.
